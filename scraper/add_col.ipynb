```python
import pandas as pd
from datetime import datetime
```

```python
file = '2026-03-12T225110'
df = pd.read_csv(f"../raw/linkedin/{file}.csv")
df.head()
```

```python
# and more column scaling but bring the date from the file name
# timestamp form ex. YYYY-MM-DD HH:MM:SS
timestamp = file.replace('T', ' ')
timestamp = datetime.strptime(timestamp, '%Y-%m-%d %H%M%S')
print(f"Timestamp from file name: {timestamp}")
```

```python
df['scraped_at'] = timestamp
```

```python
#save to csv and json
df.to_json(f"../raw/linkedin/{file}.json", orient="records", lines=True)
df.to_csv(f"../raw/linkedin/{file}.csv", index=False)
```
