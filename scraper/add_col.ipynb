```python
from datetime import datetime, timezone, timedelta
# Define the UTC+7 timezone
tz_utc_plus_7 = timezone(timedelta(hours=7))
now_utc_plus_7 = datetime.now(timezone.utc).astimezone(tz_utc_plus_7)
# timestamp = now_utc_plus_7.strftime("%Y-%m-%d %H%M%S\").astype(int)

timestamp = int(now_utc_plus_7.timestamp())
print(timestamp, type(timestamp))
# # Print the result (optional)
print(tz_utc_plus_7)
print(now_utc_plus_7)
print(timestamp)
```

```python
file = "2026-03-19T205745"
json_path = f"../raw/linkedin/{file}.json"

# File was saved as NDJSON (one JSON object per line), so use lines=True
df = pd.read_json(json_path, lines=True)

df.head()
```

```python
# If you have a column like df["scraped_at"] ="2026-03-19T205745
# You can convert it to a timestamp like this:
df["scraped_at"] = pd.to_datetime(df["scraped_at"]).astype(int) // 10**9
df.head()
```

```python
# save the updated dataframe back to json
df.to_json(f"../raw/linkedin/{file}.json", orient="records", lines=True)
```
